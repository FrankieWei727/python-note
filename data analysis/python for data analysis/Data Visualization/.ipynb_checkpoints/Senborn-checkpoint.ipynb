```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

# Distribution Plot

```python
# default distplot
sns.distplot(tips['total_bill'])
```

```python
# distplot
sns.distplot(tips['total_bill'],kde=False,bins=50)
```

```python
# default jointplot
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
# pairplot
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
# rugplot
sns.rugplot(tips['total_bill'])
```

# Categorical Plot

```python
# default barplot
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
# default countplot
sns.countplot(x='sex',data=tips)
```

```python
# default boxplot
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
# default violinplot
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
# default stripplot
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False,hue='sex',dodge=True)
```

```python
# default swarmplot
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
# default catplot, by changing the kind vaule to create different kinds of plot
sns.catplot(x='day',y='total_bill',data=tips,kind='bar')
```

# Matrix Plot

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
tc = tips.corr()
tc
```

```python
# hitmap 
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
fp = flights.pivot_table(index='month',columns='year',values='passengers')
fp
```

```python
sns.heatmap(fp)
```

```python
sns.heatmap(fp,cmap='magma',linewidths=0.5,linecolor='white')
```

```python
sns.clustermap(fp)
```

```python
sns.clustermap(fp,standard_scale=1)
```

# Grids

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')
```

# Regression Plot

```python
# default lmplot
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],
          scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',aspect=0.6,height=8)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',row='time',hue='sex')
```

# Style and Color

```python
sns.set_style('whitegrid')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
sns.despine(left=True,right=True)
```

```python
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context('poster',font_scale=2)
sns.countplot(x='sex',data=tips)
```
