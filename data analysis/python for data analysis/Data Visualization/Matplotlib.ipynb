```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2

display(x,y)
```

```python
# functional methods
plt.plot(x,y,'r-')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r-')

plt.subplot(1,2,2)
plt.plot(y,x,'b-')
```

```python
# 00
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('large plot')
axes2.plot(y,x)
axes2.set_title('small plot')
```

```python
fig, axes = plt.subplots(nrows=3,ncols=3)

# make sure the plots will not overlapping 
plt.tight_layout()
```

```python
# axes is a list of plot objects
fig, axes = plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig = plt.figure(figsize=(3,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(8,5))

axes[0].plot(x,y)
axes[1].plot(y,x)

# make sure the plots will not overlapping 
plt.tight_layout()
```

```python
# output the fig
fig.savefig('my_pic.jpg',dpi=200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x ** 2,label='X squard')
ax.plot(x,x ** 3, label = 'X cubed')

ax.legend(loc=0)
# ax.legend(loc=10)
```

## Plot Appearance 

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# set line color and style, linewidth(lw) linestyle(ls)
ax.plot(x,y, color='g',linewidth=2,alpha=0.5,linestyle='--',marker='o',markersize=5,
       markerfacecolor='r')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# set line color and style, linewidth(lw) linestyle(ls)
ax.plot(x,y, color='g',linewidth=2,linestyle='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
